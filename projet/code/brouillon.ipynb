```python
df_concat = df_concat.sample(frac=1, replace=True, ignore_index=True )
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(df_concat['Source'])
df_concat['Source']=le.transform(df_concat['Source'])
le.fit(df_concat['Destination'])
df_concat['Destination']=le.transform(df_concat['Destination'])
le.fit(df_concat['Protocol'])
df_concat['Protocol']=le.transform(df_concat['Protocol'])
le.fit(df_concat['Info'])
df_concat['Info']=le.transform(df_concat['Info'])
```

```python
df_concat
```

```python
X_train=df_concat.iloc[:100000,0:5]
X_test=df_concat.iloc[100000:,0:5]
y_train=df_concat.iloc[:100000,5:]
y_test=df_concat.iloc[100000:,5:]
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```
